```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave form')

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin,'r--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

plt.show()
```

```python
x=[2,8,10]
y=[11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
